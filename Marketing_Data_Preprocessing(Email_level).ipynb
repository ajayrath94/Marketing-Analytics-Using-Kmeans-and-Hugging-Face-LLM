```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('loan_mkt.csv')
```

```python
df2=df.copy(deep=True)
```

```python
df2
```

```python
#emails sent per email
emails_sent_per_email=df2.groupby('email_address').size().reset_index(name='count')
emails_per_email_sorted = emails_sent_per_email.sort_values(by='count', ascending=False)
emails_per_email_sorted
```

```python
#adds_open_per_email
adds_open_per_email=df2[df2['opened_flag'] == 'Open'].groupby('email_address').size().reset_index(name='count_open')
adds_open_per_email_sorted = adds_open_per_email.sort_values(by='count_open', ascending=False)
adds_open_per_email_sorted
```

```python
#adds_click_per_email
adds_click_per_email=df2[df2['clicked_flag'] == 'Click'].groupby('email_address').size().reset_index(name='count_click')
adds_click_per_email_sorted = adds_click_per_email.sort_values(by='count_click', ascending=False)
adds_click_per_email_sorted
```

```python
#adds_bounce_per_email
adds_bounce_per_email=df2[df2['bounce_flag'] == 'Bounce'].groupby('email_address').size().reset_index(name='count_bounce')
adds_bounce_per_email_sorted = adds_bounce_per_email.sort_values(by='count_bounce', ascending=False)
adds_bounce_per_email_sorted
```

```python
result_df = pd.merge(emails_per_email_sorted, adds_open_per_email_sorted, on='email_address', how='left')
result_df
```

```python
result_df2 = pd.merge(result_df, adds_click_per_email_sorted, on='email_address', how='left')
result_df2
```

```python
result_df3 = pd.merge(result_df2, adds_bounce_per_email_sorted, on='email_address', how='left')
result_df3
```

```python
#send_id_per_email_address
send_id_per_email=df2.groupby('email_address')['send_id'].nunique().reset_index(name='unique_send_count')
send_id_per_email_sorted=send_id_per_email.sort_values(by='unique_send_count', ascending=False)
send_id_per_email_sorted
```

```python
result_df4 = pd.merge(result_df3, send_id_per_email_sorted, on='email_address', how='left')
result_df4
```

```python
#application_ids per email_address
apps_per_email=uq_app_df.groupby('email_address').size().reset_index(name='apps_count')
apps_per_email_sorted = apps_per_email.sort_values(by='apps_count', ascending=False)
apps_per_email_sorted
```

```python
df5=uq_app_df[['email_address','application_instance_id','created_date']]
df5
```

```python
#applications_assocaited_with_email
result_df6 = pd.merge(result_df5, df5, on='email_address', how='left')
result_df6
```

```python
#Emails with one or more application_instance_id
filtered_df = result_df6[result_df6['application_instance_id'].notna()]
filtered_df
```

```python
#emails with no applications
filtered_df2 = result_df6[result_df6['application_instance_id'].isna()]
filtered_df2
```

```python
filtered_df['rank'] = filtered_df.sort_values('created_date', ascending=False)\
              .groupby('email_address')\
              .cumcount() + 1
```

```python
#emails_with_most_recent_application based on created date
filtered_df = filtered_df[filtered_df['rank'] == 1].copy()
filtered_df
filtered_df.drop('rank',axis=1,inplace=True)
filtered_df

```

```python
#email_address with recent_application
final_df = pd.concat([filtered_df, filtered_df2], axis=0, ignore_index=True)
final_df
```
