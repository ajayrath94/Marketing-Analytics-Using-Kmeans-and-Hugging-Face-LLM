```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
```

```python
df=pd.read_csv('Email_Level_Marketing_Analytics.csv')
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
sentence_df=df.copy(deep=True)
```

```python
sentence_df.fillna('unknown',inplace=True)
sentence_df
```

```python
def compile_text(x):


    text =  f"""Contact_Count: {x['count']},  
                Open_Count: {x['count_open']}, 
                Click_Count: {x['count_click']}, 
                Bounce_Count: {x['count_bounce']}, 
                Campagin_Count: {x['unique_send_count']}, 
                Apps_Count: {x['apps_count']}, 
                Fico: {x['fico']}, 
                Credit_score: {x['credit_score']},
                Liability_Bankruptcy_count: {x['liability_bankruptcy_count']},
                Open%: {x['open%']},
                Click%: {x['click%']},
                Bounce%: {x['bounce%']},
                Latest_Status: {x['latest_status']},
                Race: {x['Race']},
                Women_Owned: {x['Women Owned']},
                Veteran: {x['Veteran']},
                Rural: {x['Rural']},
                Ethnicity: {x['Ethnicity']},
                State: {x['state']},
                Minority: {x['minority']},
                Total_Employees: {x['Total Employees']},
                County: {x['County']},
                City: {x['City']},
                Congressional_District: {x['Congressional District']},
                NAICS_National_Industry_Title: {x['NAICS National Industry Title']},
                LMI: {x['lmi']},
                Majority_Minority_census: {x['majority_minority_census']}
                
                
                
            """

    return text
```

```python
sentences = sentence_df.apply(lambda x: compile_text(x), axis=1).tolist()
```

```python
model = SentenceTransformer(r"sentence-transformers/paraphrase-MiniLM-L6-v2")
```

```python
output = model.encode(sentences=sentences, show_progress_bar= True, normalize_embeddings  = True)
```

```python
df_embedding = pd.DataFrame(output)
df_embedding


```

```python
df_embedding.to_csv("embedding_train.csv",index = False)
```
